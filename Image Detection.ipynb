```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import cv2
import random

DATADIR="dataset"
CATEGORIES=["vinod","mohsin","rick"]
```

```python
data=[]
for category in CATEGORIES:
    path=os.path.join(DATADIR,category)
    label=CATEGORIES.index(category)
    
    
    for img in os.listdir(path):
        img=os.path.join(path,img)
        img_arr=cv2.imread(img)
        img_arr=cv2.resize(img_arr,(100,100))
         
        data.append([img_arr,label])
```

```python
data
```

```python
random.shuffle(data)
```

```python
X=[]
y=[]

for features,label in data:
    X.append(features)
    y.append(label)
```

```python
X=np.array(X)
y=np.array(y)
```

```python
X=X/255
```

```python
X
```

```python
X.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout,Activation
```

```python
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3,3),input_shape=(100, 100, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10, activation='sigmoid'))


model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=100,validation_split=0.1)
```

```python
model.summary()
```

```python
import keras.utils as image

img_pred=image.load_img("dataset\\vinod\\1.jpg",target_size=(100,100))

img_pred=image.img_to_array(img_pred)
img_pred=np.expand_dims(img_pred,axis=0)

result=model.predict(img_pred)

print(result)
if result[0][0]>result[0][1]:
    prediction="rick"
else:
    prediction="vinod"
    
print(prediction)
```

```python
# save model and architecture to single file
model.save("model.h5")
print("Saved model to disk")
```
